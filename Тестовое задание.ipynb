```python
import pandas as pd 
```

```python
SMS_df = pd.read_csv('SMS.csv', sep=';')
```

```python
SMS_df
```

```python
SMS_df.columns = ['Action_time', 'CustomerID','Price','OrderID']
```

```python
promocode_df = pd.read_excel('promocode.xlsx')
```

```python
promocode_df.columns = ['OrderID', 'Creationdate', 'Price_with_deliver',
        'ClientID', 'PromoCode']
```

```python
promocode_df
```

```python
promocode_df['PromoCode'].unique()
```

```python
# По OrderID
Clients_df = SMS_df.merge(promocode_df, how='outer' )
```

```python
Clients_df
```

```python
# NaN == NaN == False
task1_df = Clients_df.query('PromoCode == PromoCode and Action_time == Action_time')
```

```python
task1_df
```

```python
#NaN отсутствуют
task1_df.isna().apply('sum')
```

```python
# Люди получившие промокод и SMS
task1_df.ClientID.unique()
```

```python
# Промокод и SMS
users_task1 = task1_df.CustomerID.nunique()
print('Уникальные пользователи', users_task1)
ravenue_task1 = sum(task1_df.Price_with_deliver)
print('Выручка',ravenue_task1)
print('Уникальные заказы',task1_df.OrderID.nunique())
ARPU_task1 = ravenue_task1 / users_task1 
print('ARPU', ARPU_task1)
```

```python
# Кто получил только промокод 
task2_df = Clients_df.query('PromoCode == PromoCode and Action_time != Action_time')
```

```python
task2_df
```

```python
# Только промокод, без СМС
users_task2 = task2_df.ClientID.nunique()
print('Уникальные пользователи', users_task2)
ravenue_task2 = sum(task2_df.Price_with_deliver)
print('Выручка', ravenue_task2)
print('Уникальные заказы',task2_df.OrderID.nunique())
ARPU_task2 = ravenue_task2 / users_task2 
print('ARPU', ARPU_task2)
```

```python
task3_df = Clients_df.query('PromoCode != PromoCode and Action_time == Action_time')
```

```python
task3_df
```

```python
#Подготовка колонки Price
task3_profit = task3_df.Price.apply(lambda x: float(x.replace(',','.')))
```

```python
#Только SMS
users_task3 = task3_df.CustomerID.nunique()
print('Уникальные пользователи', users_task3)
ravenue_task3 = sum(task3_profit)
print('Выручка',ravenue_task3)
print('Уникальные заказы',task3_df.OrderID.nunique())
ARPU_task3 = ravenue_task3 / users_task3 
print('ARPU', ARPU_task3)
```
